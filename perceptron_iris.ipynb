```python
X=data.drop('Class',axis=1)
Y=data['Class']
```

```python
from sklearn.metrics import accuracy_score
```

```python
class Perceptron:
    def __init__(self):
        self.w=None
        self.b=None
    
    def model(self,x):
        return 1 if (np.dot (self.w,x) >= self.b) else 0

    
    def predict(self,X):
        Y = []
        for x in X:
          result = self.model(x)
          Y.append(result)
        return np.array(Y)

    def fit(self,X,Y,epochs=1):
        self.w=np.ones(X.shape[1])
        self.b = 0
        accuracy = {}

        for i  in range (epochs):
            for x,y in zip(X,Y):
                y_pred = self.model(x)
                if y==1 and y_pred==0:
                    self.w = self.w + x
                    self.b = self.b + 1
                elif y==0 and y_pred==1:
                    self.w = self.w - x
                    self.b = self.b - 1
        

            accuracy[i] = accuracy_score(self.predict(X),Y) #np.sum(y_pred==y)/y.shape[0]

        plt.plot(accuracy.values())
        plt.xlable('Epochs')
        plt.ylable('Accuracy')
        plt.show()
            
```

```python
perceptron = Perceptron()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def sigmoid(x,w=2.0,b=-1.0):
    return 1 / (1 + np.exp(w*x+b))

# Generate x values
x = np.linspace(-10, 10, 1000)

y = sigmoid(x)


plt.figure(figsize=(8, 4))
plt.plot(x, y, label='Sigmoid Function', color='blue')
plt.title('Sigmoid Function')
plt.xlabel('x')
plt.ylabel('σ(x)')
plt.grid(True)
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.50)
plt.legend()
plt.show()
```
